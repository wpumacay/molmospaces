```python
from pathlib import Path
import json
import pandas as pd


benchmark_dir = Path("/weka/prior/datasets/robomolmo/bench_v2/")
rows = []

for bench_fn in filter(Path.is_dir, benchmark_dir.iterdir()):
    bench_name = bench_fn.name
    # in 1ep, 10ep, 200ep in bench_name, skip
    if any(ep in bench_name for ep in ["1ep", "10ep", "200ep"]):
        continue
    bench_metadata = bench_fn / "benchmark_metadata.json"
    if not bench_metadata.exists():
        print(f"Missing metadata for {bench_name}")
        continue
    with bench_metadata.open() as f:    
        meta = json.loads(f.read())

    rows.append({
        "name": bench_name,
        "num_episodes": meta["num_episodes"],
        "num_houses": meta["num_houses"],
        "num_categories": len(meta["object_category_counts"]),
    })

df = pd.DataFrame(rows)
print(df.to_latex(index=False, escape=True))
print(df)
```
